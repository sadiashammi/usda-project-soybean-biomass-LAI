```python
# LR MODELLING
# Import Module
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
import statsmodels.api as sm

from sklearn.model_selection import train_test_split
#column name as indexes
index=['CC','NDVI', 'GNDVI', 'EVI2', 'NDRE', 'ARVI', 'CCCI', 'GRRI', 'CARI', 'NAVI','SCCCI', 'CIRE', 'CVI', 'GCVI']
for i in index:
    #print(i)

    data=data
    x=np.array(data[[i]].values)
    y=np.array(data['column name'].values).reshape(-1,1)

    model=LinearRegression()
    model.fit(x,y)
    y_pred=model.predict(x)

    metrics_dict = metrics.r2_score(y, y_pred) #(actual, predicted)
    #print(metrics_dict)
    r2=np.sqrt(metrics.mean_squared_error(y, y_pred))
    r1= metrics_dict
    res=[]
    res.append(i)
    res.append(r1)
    res.append(r2)
    
    print(res)
```
